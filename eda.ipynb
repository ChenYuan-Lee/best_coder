```python
# Compute Correlation between columns
target_column = "Dis"
for column in data:
    print(f'Correlation between {column} and target is {data[target_column].corr(data[column])}')
```

```python
# Find out missing data percentage for columns which are missing data
for column in data:
    num_missing_data = data[column].isna().sum()
    len_data =len(data)
    if num_missing_data:
        print(f'Mising data rate for {column} is {(num_missing_data/len_data*100).round(2)}%')
```
